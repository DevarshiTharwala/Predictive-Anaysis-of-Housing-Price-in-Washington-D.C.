```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from factor_analyzer import FactorAnalyzer
```

```python
df2 = pd.read_csv(r'C:\DEV\OSU Sem_4\Numerodata\Data Science\Project1\datasetopenall.csv')
df2.head()
```

```python
X = df2.iloc[:, :-1]
y = df2.iloc[:, 278]
```

```python
y.head()
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X)

X = scaler.transform(X)

X = pd.DataFrame(data = X)
X.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=128)
principalComponents = pca.fit_transform(X)
XPCA = pd.DataFrame(data = principalComponents)
```

```python
XPCA.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
XPCA_train, XPCA_test, y_train, y_test = train_test_split(XPCA, y, test_size = 0.3)
```

```python
XPCA_train.shape, XPCA_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(XPCA_train, y_train)
```

```python
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_, XPCA.columns, columns = ['Coefficient'])
```

```python
coeff_df
```

```python
predictions = lm.predict(XPCA_test)
```

```python
plt.scatter(y_test, predictions)
```

```python
sns.distplot((y_test-predictions), bins = 50)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
```

```python
print('MSE:', metrics.mean_squared_error(y_test, predictions))
```

```python
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
from sklearn.metrics import r2_score
```

```python
print('R2 Value: ', r2_score(y_test, predictions))
```
